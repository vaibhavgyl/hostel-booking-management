```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("Hotels Booking Data (1).csv")
df.head()
```

```python
df['booking_date'] = pd.to_datetime(df['booking_date'])
df['check_in_date'] = pd.to_datetime(df['check_in_date'])
```

```python
# Count bookings by stay type
stay_type_counts = df['stay_type'].value_counts()

# Calculate percentage
stay_type_percentage = (stay_type_counts / stay_type_counts.sum()) * 100

# Display results
stay_type_summary = pd.DataFrame({
    'Bookings': stay_type_counts,
    'Percentage (%)': stay_type_percentage.round(2)
})

stay_type_summary
plt.figure(figsize=(6,4))
stay_type_counts.plot(kind='bar', color=['#1f77b4', '#ff7f0e'])

plt.title("Leisure vs Business Bookings")
plt.xlabel("Stay Type")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
## generating for the trend 2 
```

```python
# Count bookings by booking channel
channel_counts = df['booking_channel'].value_counts()

# Calculate percentage share
channel_percentage = (channel_counts / channel_counts.sum()) * 100

# Create summary table
channel_summary = pd.DataFrame({
    'Bookings': channel_counts,
    'Percentage (%)': channel_percentage.round(2)
})

channel_summary
plt.figure(figsize=(7,4))
channel_counts.plot(
    kind='bar',
    color=['#2ca02c', '#1f77b4', '#ff7f0e']
)

plt.title("Bookings by Channel")
plt.xlabel("Booking Channel")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
plt.figure(figsize=(6,6))
channel_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Booking Channel Share")
plt.ylabel("")
plt.show()
```

```python
## generating chart for trend 3
```

```python
# Count refund status
refund_counts = df['refund_status'].value_counts()

# Calculate percentage
refund_percentage = (refund_counts / refund_counts.sum()) * 100

# Create summary table
refund_summary = pd.DataFrame({
    'Bookings': refund_counts,
    'Percentage (%)': refund_percentage.round(2)
})

refund_summary
refund_by_channel = pd.crosstab(df['booking_channel'], df['refund_status'])

refund_by_channel.plot(
    kind='bar',
    stacked=True,
    figsize=(7,4)
)

plt.title("Refunds by Booking Channel")
plt.xlabel("Booking Channel")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=0)
plt.legend(title="Refund Status")
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
## Booking Lead Time
```

```python
import pandas as pd

# Convert dates to datetime
df['booking_date'] = pd.to_datetime(df['booking_date'], dayfirst=True)
df['check_in_date'] = pd.to_datetime(df['check_in_date'], dayfirst=True)

# Calculate lead time in days
df['lead_time_days'] = (df['check_in_date'] - df['booking_date']).dt.days

# Preview
df[['booking_date', 'check_in_date', 'lead_time_days', 'refund_status']].head()
```

```python
# Define lead time bins
bins = [0, 7, 14, 30, 60, 90, 180, 365, 1000]
labels = ['0-7', '8-14', '15-30', '31-60', '61-90', '91-180', '181-365', '365+']

df['lead_time_bin'] = pd.cut(df['lead_time_days'], bins=bins, labels=labels, right=False)

# Preview
df[['lead_time_days', 'lead_time_bin', 'refund_status']].head()
```

```python
# Cross-tab: Lead Time Bin vs Refund Status
refund_leadtime = pd.crosstab(df['lead_time_bin'], df['refund_status'])

refund_leadtime
```

```python
# Plot stacked bar chart
refund_leadtime.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5),
    colormap='Set3'
)

plt.title("Cancellations / Refunds by Booking Lead Time")
plt.xlabel("Lead Time (Days)")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title="Refund Status")
plt.show()
```

```python
## stay length
```

```python


# Ensure dates are datetime
df['check_in_date'] = pd.to_datetime(df['check_in_date'], dayfirst=True)
df['check_out_date'] = pd.to_datetime(df['check_out_date'], dayfirst=True)

# Calculate stay length in days
df['stay_length'] = (df['check_out_date'] - df['check_in_date']).dt.days

# Preview
df[['check_in_date', 'check_out_date', 'stay_length', 'refund_status']].head()
```

```python
# Define stay length bins
bins = [0, 1, 3, 5, 7, 10, 15, 30, 100]
labels = ['1', '2-3', '4-5', '6-7', '8-10', '11-15', '16-30', '30+']

df['stay_length_bin'] = pd.cut(df['stay_length'], bins=bins, labels=labels, right=True)

# Preview
df[['stay_length', 'stay_length_bin', 'refund_status']].head()
```

```python
# Cross-tab: Stay Length Bin vs Refund Status
refund_stay = pd.crosstab(df['stay_length_bin'], df['refund_status'])

refund_stay
```

```python


# Plot stacked bar chart
refund_stay.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5),
    colormap='Paired'
)

plt.title("Cancellations / Refunds by Stay Length")
plt.xlabel("Stay Length (Nights)")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title="Refund Status")
plt.show()
```
